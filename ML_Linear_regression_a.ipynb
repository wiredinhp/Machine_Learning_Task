```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('data.csv',index_col=0)
print(df)
```

```python
plt.plot(df)
plt.show()
```

```python
matrix = np.array(df.values,'float')
```

```python
y=[]
for i in range(10):
    for j in range(10):
        y.append(matrix[j,i])
print(y)
    
```

```python
X=[]
for i in range(100):
    X.extend([i])
X= np.array(X)
X = (X/100.00)
print(X)
```

```python
#Initialization
m = 100
X = X.reshape([100,1])
x = np.hstack([np.ones_like(X),X])
theta = np.zeros([2,1])
print(x)
print(theta)
```

```python
def cost(x,y,theta):
    
    a = 1/100.00
    b = np.sum(abs(((x@theta)-y)**3))
    j = (a)*(b)
    return j
print(cost(x,y,theta))
```

```python
def gradient(x,y,theta):
    
    alpha = 0.001
    iteration = 500
    m=100.00
#gradient descend algorithm
    for iter in range(500):
        error = ((x @ theta) -y)
        temp0 = theta[0] - ((alpha/m) * np.sum(((error**2)*(np.sign(error)))*x[:,0]))
        temp1 = theta[1] - ((alpha/m) * np.sum(((error**2)*(np.sign(error)))*x[:,1]))
        theta[0]=temp0
        theta[1]=temp1
    J = (1 / m ) * np.sum(abs(((x@theta)-y)**3))
    return theta,J
#Lets check the values
theta, J = gradient(x,y,theta)
print(J)
print(theta)
```

```python
x_plot = x[:,1]
y_plot = y
x_line = x[:,1]
y_line = theta[0]+((theta[1])*x_line)
plt.plot(x_plot,y_plot)
plt.plot(x_line,y_line)
plt.show()
```

```python
#computing theta values
theta0_vals = [[0.000], [0.001], [0.002], [0.003], [0.004], [0.005], [0.006], [0.007], [0.008], [0.009]]
theta1_vals = [[0.000], [0.001], [0.002], [0.003], [0.004], [0.005], [0.006], [0.007], [0.008], [0.009]]
theta0_vals= np.array(theta0_vals)
theta1_vals= np.array(theta1_vals)
#computing J_vals
J_vals = np.zeros([np.size(theta0_vals),np.size(theta1_vals)])
for i in range(0,np.size(theta0_vals)):
    for j in range(0,np.size(theta1_vals)):
        t = [theta0_vals[i,:],theta1_vals[j,:]]
        J_vals[i,j] = cost(x,y,t)
print(J_vals)
```

```python
from mpl_toolkits.mplot3d import Axes3D 
fig = plt.figure()
ax = fig.add_subplot(projection ='3d')
ax.plot_surface(theta0_vals,theta1_vals,J_vals, edgecolor ='green')
ax.set_xlabel('theta0')
ax.set_ylabel('theta1')
ax.set_zlabel('J_vals')
plt.show()
```
