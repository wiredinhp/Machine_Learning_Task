```python
X=[]
for i in range(100):
    X.extend([i])
X= np.array(X)
X = (X/100.00)
print(X)
```

```python
#Initialization
m = 100
X = X.reshape([100,1])
x = np.hstack([np.ones_like(X),X,X**2])
theta = np.zeros([3,1])
print(x)
print(theta)
```

```python
def cost(x,y,theta):
    
    a = 1/100.00
    b = np.sum(abs(((x@theta)-y)**7))
    j = (a)*(b)
    return j
print(cost(x,y,theta))
```

```python
# running various values for one iteration just to check alpha
def gradient(x,y,theta):
    
    alpha = 0.000000001
    iteration = 2000
    m=100.00
#gradient descend algorithm
    for iter in range(2000):
        
        error = ((x @ theta) -y)
        temp0 = theta[0] - ((alpha/m) * np.sum(((np.sign(error))*(error**6))*x[:,0]))
        temp1 = theta[1] - ((alpha/m) * np.sum(((np.sign(error))*(error**6))*x[:,1]))
        temp2 = theta[2] - ((alpha/m) * np.sum(((np.sign(error))*(error**6))*x[:,2]))
        theta[0]=temp0
        theta[1]=temp1
        theta[2]=temp2
    J = (1 / m ) * np.sum(abs(((x@theta)-y)**7))
    return theta,J
#Lets check the values
theta, J = gradient(x,y,theta)
print(J)
print(theta)
# after running so many small values (like hell lot of small values) we can infer that J is min around 0,0,0
```
